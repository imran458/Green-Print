```python

def cellphoneEmissions():
    mAh = int(input("How many mAh does your phone require? "))
    Volt = int(input("How many volts is your charger? "))
    frequency = int(input("How often do you charge per day? "))
    
    #formula to convert from mAh to watt hours(wh)
    
    wh = (mAh * Volt)/1000
    
    #formula for killowatt hour
    
    kh = wh/1000
    
    #final calculation for carbon dioxide emissions
    carbon_emissions_per_charge = (kh * 1.6)
    
    final_carbon_emissions = carbon_emissions_per_charge * frequency 
    
    daily_carbon = final_carbon_emissions + 0
    weekly_carbon_emissions = final_carbon_emissions * 7
    monthly_carbon_emissions = final_carbon_emissions * 30
    
    yearly_carbon_emissions = final_carbon_emissions * 365
   
    
    print("Your phone produces ",daily_carbon,"pounds of CO2 per day")
    print("Your phone produces ",weekly_carbon_emissions,"pounds of CO2  per week")
    print("Your phone produces ",monthly_carbon_emissions,"pounds of CO2  per month")
    print("Your phone produces ",yearly_carbon_emissions,"pounds of CO2  per year")
    

cellphoneEmissions()
    
```
